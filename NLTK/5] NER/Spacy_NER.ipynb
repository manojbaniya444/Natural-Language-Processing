```python
import spacy
```

```python
!python -m spacy download en_core_web_md


nlp = spacy.load("en_core_web_md")
```

```python
sentence = "Mero iPhone 13 ko price 130000 ho ra tshirt chai Antartica baata kineko"
```

```python
doc = nlp(sentence)
```

```python
doc
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
